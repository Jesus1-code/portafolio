DATA WRANGLING LIMPIEZA Y PREPARACIÓN DE DATOS

```python
# Conversión de tipos de datos: La columna 'Date' está en formarto string, la convertiremos a datetime.
df['Date'] = pd.to_datetime(df['Date'], format='%d/%m/%Y')   # convierte a formato fecha.
```

```python
# Crear nuevas columnas para el análisis: Total de goles por partido, resultado detallado, etc.
df["TotalGoals"] = df["FTHG"] + df["FTAG"]   # suma goles locales + visitantes.

df["ResultDetailed"] = np.where(df["FTHG"] > df["FTAG"], "Home Win",
                                 np.where(df["FTHG"] < df["FTAG"], "Away Win", "Draw"))
```

```python
# Filtrados de datos: si asumimos que queremos solo partidos completos (excluimos futuros si hay):
df = df[df["FTR"].notnull()]   # filtra solo filas con resultados final no nulo.
```

```python
# Agrupación por equipos: se crea DataFrames agregados para análisis posteriores.
home_stats = df.groupby("HomeTeam").agg({"FTHG": "sum", "FTAG": "sum", "HS": "mean", "HST": "mean", "HC": "mean"}).reset_index()
home_stats.rename(columns={"FTHG": "HomeGoalsScored", "FTAG": "HomeGoalsConceded"}, inplace=True)   # renombrar para mejor claridad.

away_stats = df.groupby("AwayTeam").agg({"FTAG": "sum", "FTHG": "sum", "AS": "mean", "AST": "mean", "AC": "mean"}).reset_index()
away_stats.rename(columns={"FTAG": "AwayGoalsScored", "FTHG": "AwayGoalsConceded"}, inplace=True)   # renombrar para mejor claridad.
```

```python
# Unión de stats locales y visitantes para stats totales por equipo.
team_stats =pd.merge(home_stats, away_stats, left_on="HomeTeam", right_on="AwayTeam")
team_stats["Team"] = team_stats["HomeTeam"]   # crea columna de equipo unificada.
team_stats["TotalGoalsScored"] = team_stats["HomeGoalsScored"] + team_stats["AwayGoalsScored"]
team_stats["TotalGoalsConceded"] = team_stats["HomeGoalsConceded"] + team_stats["AwayGoalsConceded"]
team_stats = team_stats[["Team", "TotalGoalsScored", "TotalGoalsConceded"]]   # seleccionar columnas relevantes.

print("\nData Wrangling completado. Nuevas columnas agregadas: TotalGoals, ResultDetailed.")
```

```python
# Función para detectar y manejar outliers en una columna específica.
def handle_outliers(df, col):
  Q1 = df[col].quantile(0.25)   # primer cuartil.
  Q3 = df[col].quantile(0.75)   # tercer cuartil.
  IQR = Q3 - Q1   # rango intercuartilico.
  lower_bound = Q1 - 1.5 * IQR   # límite inferior.
  upper_bound = Q3 + 1.5 * IQR   # límite superior.

  # Identificar outliers
  outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]
  print(f"\nOutliers detectados en {col}: {len(outliers)}")
  if not outliers.empty:
    print(outliers[["Date", "HomeTeam", "AwayTeam", col]])   # muestra los outliers para la inspección.

  return lower_bound, upper_bound # Return the bounds

# Detectar outliers for relevant columns
columns_to_check = ["TotalGoals", "HS", "AS"]

for col in columns_to_check:
  lower_bound, upper_bound = handle_outliers(df, col) # Get the bounds

  # Visualización: Boxplot antes de manejar outliers for the current column.
  plt.figure(figsize=(8, 5))
  sns.boxplot(x=df[col])
  plt.title(f"Boxplot de {col} (Antes de manejo de outliers)", fontsize=14)
  plt.savefig(f"boxplot_{col}_before.png")   # guarda imagen.
  plt.show()

  # Manejo: Capear los outliers a los límites (reemplazar valores extremos por los bounds).
  df[col] = df[col].clip(lower= lower_bound, upper= upper_bound)   # capear valores.
  print(f"\nOutliers han sido manejados en {col}.")

  # Boxplot después.
  plt.figure(figsize=(8, 5))
  sns.boxplot(x=df[col])
  plt.title(f"Boxplot de {col} (Después de manejo de outliers)", fontsize=14)
  plt.savefig(f"boxplot_{col}_after.png")   # guarda imagen.
  plt.show()

print("\nManejo de Outliers completado. Los valores extremos han sido capeados para evitar sesgos en el análisis posterior.")
```

ANÁLISIS EXPLORATORIO DE DATOS (EDA) CON VISUALIZACIONES.

```python
# Configuración de un estilo para los gráficos con seaborn.
sns.set(style="whitegrid", palette="muted")   # estilo moderno y colores suaves.
plt.style.use("seaborn-v0_8-darkgrid")

# Visualización 1: Distribución de resultados, ¿Cuántos partidos terminan en victoria local, visitante o empate?

plt.figure(figsize=(8, 6))   # tamaño del gráfico.
ax=sns.countplot(data=df, x="ResultDetailed", order=["Home Win", "Draw", "Away Win"])   # gráfico de barras.
ax.bar_label(ax.containers[0])   # etiquetas de barras.
plt.title("Distribución de Resultados en La Liga 2024-2025", fontsize=16)   # titulo del gráfico.
plt.xlabel("Resultado", fontsize=12)   # etiqueta eje x.
plt.ylabel("Número de partidos", fontsize=12)   # etiqueta eje y.
plt.savefig("result_distribution.png")   # guarda la imagen para github.
plt.show()   # muestra el gráfico.
```

```python
# Visualización 2: Distribución de goles totales por partido, ¿Son los partidos de alta o baja anotación?
plt.figure(figsize=(10, 6))
sns.histplot(df["TotalGoals"], bins=10, kde=True)   # histograma con curva de densidad.
plt.title("Distribución de goles totales por partido", fontsize=16)   # título del gráfico.
plt.xlabel("Total de goles", fontsize=12)   # etiqueta eje x.
plt.ylabel("Frecuencia", fontsize=12)   # etiqueta eje y.
plt.savefig("goal_distribution.png")    # guarda imagen.
plt.show()   # muestra el gráfico
```

```python
# Visualizacipón 3: Rendimiento por equipo, goles anotados vs concedidos. ¿Quiénes son los equipos más fuertes?
plt.figure(figsize=(12, 8))   # tamaño grande para scatter plot.
sns.scatterplot(data=team_stats, x="TotalGoalsScored", y="TotalGoalsConceded", hue="Team", s=100)   # scatter con colores por equipo.
plt.title('Goles Anotados vs Concedidos por Equipo', fontsize=16)  # título del gráfico.
plt.xlabel('Goles Anotados', fontsize=12)  # etiqueta eje x.
plt.ylabel('Goles Concedidos', fontsize=12)  # etiqueta eje y.
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')  # leyenda fuera del gráfico.
plt.savefig('team_performance.png')  # guarda la imagen.
plt.show()  # muestra el gráfico.
```

```python
# Visualización 4: Evolución de goles a lo largo de la temporada, ¿Aumentan los goles hacia el final de la temporada?
df["Month"] = df["Date"].dt.to_period("M")   #extrae mes para agrupación temporal.
monthly_goals = df.groupby("Month")["TotalGoals"].mean().reset_index()   #promedio de goles por mes.
monthly_goals["Month"] = monthly_goals["Month"].astype(str)   #convierte a string para plotting.
plt.figure(figsize=(10, 6))  # tamaño del gráfico.
sns.lineplot(data=monthly_goals, x='Month', y='TotalGoals', marker='o')  # línea con marcadores.
plt.title('Promedio de Goles por Mes en la Temporada', fontsize=16)  # título.
plt.xlabel('Mes', fontsize=12)  # etiqueta eje x.
plt.ylabel('Promedio de Goles', fontsize=12)  # etiqueta eje y.
plt.xticks(rotation=45)  # rota etiquetas para legibilidad.
plt.savefig('monthly_goals.png')  # guarda imagen.
plt.show()  # muestra el gráfico.
```

```python
# Visualizacón 5:  Relación entre Goles y Disparos (Matriz de Correlación), ¿Qué relación existe entre los goles marcados y los tiros al arco?
# Calcular la matriz de correlación entre las variables relevantes.
correlation_df = df[["FTHG", "HST", "FTAG", "AST"]].corr()   #calcula la correlación de Pearson.

plt.figure(figsize=(8, 6))   # tamaño del gráfico.
sns.heatmap(correlation_df, annot=True, cmap="coolwarm", fmt=".2f", linewidths=.5)   # genera un mapa de calor, "annot=True" muestra los valores.
plt.title('Matriz de Correlación entre Variables de Goles y Disparos', fontsize=16)   # título del gráfico.
plt.savefig('correlation_matrix.png')   # guarda imagen.
plt.show()   # muestra el gráfico.
```

```python
# Visualización 6: Distribución de tarjetas por equipo, ¿Qué equipos son los más agresivos en cuanto a tarjetas apersividas?

plt.figure(figsize=(12, 6))   # tamaño del gráfico.
plt.suptitle("Distribución de tarjetas amarillas y rojas", fontsize=16)

plt.subplot(1, 2, 1)
sns.boxplot(data=df[["HY", "AY"]], palette='YlOrRd')   # crea un diagrama de caja que muestra la distribución de los datos.
plt.title("Tarjetas Amarillas", fontsize=14)
plt.xlabel("Tipo de Equipo", fontsize=12)
plt.ylabel("Número de tarjetas", fontsize=12)
plt.xticks(ticks=[0, 1], labels=["Local", "Visitante"])

plt.subplot(1, 2, 2)
sns.boxplot(data=df[["HR", "AR"]], palette='YlOrRd')
plt.title("Tarjetas Rojas", fontsize=14)
plt.xlabel("Tipo de Equipo", fontsize=12)
plt.ylabel("Número de tarjetas", fontsize=12)
plt.xticks(ticks=[0, 1], labels=["Local", "Visitante"])

plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.savefig("yellow_red_cards.png")   # guarda la imagen.
plt.show()    # muestra el gráfico
```
